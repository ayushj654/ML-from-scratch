```python
import pandas as pd
import numpy as np
from sklearn import datasets
import scipy as sc
```

```python
class MyLRGD:

    def __init__(self,learning_rate=0.3,max_iterations=100) -> None:
        self.n = learning_rate
        self.iter = max_iterations
        self.W = None

    def intialize_weights(self,input_data):
        return np.random.uniform(0,1,input_data.shape[1])

    def fit(self,X,y):
        X_adj = np.insert(X.to_numpy(),0,1,axis=1)
        self.W = self.intialize_weights(X_adj)
        for i in range(self.iter):
            y_pred = X_adj.dot(self.W)
            error = 0.5*(y-y_pred)**2
            grad = error.dot(X_adj)
            print(self.W, grad)
            self.W -= self.n*grad

    def transform(self,X):
        pass
```
